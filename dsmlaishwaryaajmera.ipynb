```python
#I WANT TO FILL MISSING VALUES WITH MEAN 
#first i want to convert colums with units to numeric
columns_to_clean = ['mileage','engine','max_power', 'torque']
def extract_numeric(series):
    return series.str.extract(r'(\d+\.\d+|\d+)').astype(float)
for column in columns_to_clean:
    car_data[column]= extract_numeric(car_data[column].astype(str))
print(car_data[columns_to_clean])
columns_to_fill=['mileage','engine','max_power','torque','seats']
for columns in columns_to_fill:
    car_data[column]=car_data[column].fillna(car_data[column].mean())
print(car_data[columns_to_fill])    
```
